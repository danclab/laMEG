# Tutorial 08 - Sliding layer EBB

This tutorial demonstrates the need for a sliding window layer_EBB reconstruction to accurately infer two distinct laminar sources across time and space. It builds on the laminar CSD tutorial and addresses limitations, as well as best practices for using EBB for laminar inference.

It is composed of two parts: one adressing the need for sliding windows and the other for a ebb_layer implementation in the case of correlated sources across layers.

Two sources (temporal gaussian function) are simulated in superficial and deep layers, with varying temporal jitter (succeeding or simultaneous). Source reconstruction is performed on the simulated sensor data using a forward model based on the multilayer mesh, first on the whole time window using EBB, then using a sliding window EBB, an finally using a modified version of EBB with specific priors accounting for correlated sources across layers. Laminar CSD analysis is then ran on the laminar source reconstructed signals at the location of the peak variance.

Now we can compute and plot the laminar CSD

```python
# Set a function for loading the time series, computing CSD and plotting
def load_and_plot_csd(sim_fname, surf_set, layer_verts, layer_dists, method, MU=None):
    # Get source time series for each layer (already averaged)
    if MU is not None:
        mean_layer_ts, time, _ = load_source_time_series(sim_fname, mu_matrix=MU, vertices=layer_verts)
    else: 
        mean_layer_ts, time, _ = load_source_time_series(sim_fname, vertices=layer_verts)
        
    if method == 'kcsd':
        csd = compute_csd(mean_layer_ts, np.sum(layer_dists), s_rate, method='KCSD1D')
    if method == 'standard_smooth':
        csd = compute_csd(mean_layer_ts, np.sum(layer_dists), s_rate, method='StandardCSD', smoothing='cubic')

    col_r = plt.cm.cool(np.linspace(0,1, num=surf_set.n_layers))
    plt.figure(figsize=(15,4))
    plt.subplot(1,3,1)
    for l in range(surf_set.n_layers):
        plt.plot(time, mean_layer_ts[l,:], label=f'{l}', color=col_r[l,:])
    plt.legend(loc='upper left')
    plt.xlabel('Time (ms)')
    plt.ylabel('Source (nAm)')

    plt.subplot(1,3,2)
    col_r = plt.cm.cool(np.linspace(0,1, num=csd.shape[0]))
    for l in range(csd.shape[0]):
        plt.plot(time, csd[l,:], color=col_r[l,:])
    plt.xlabel('Time (ms)')
    plt.ylabel('CSD')

    ax=plt.subplot(1,3,3)
    plot_csd(csd, time, ax, n_layers=surf_set.n_layers)
    plt.xlabel('Time (ms)')
    plt.ylabel('Layer')
    plt.tight_layout()
```

```python
load_and_plot_csd(sim_fname, surf_set, layer_verts, layer_dists, 'kcsd', MU=MU)
```

#### Invert on a sliding window

We then reconstruct layer source time series using a sliding time window of 25ms.

```python
# uses a modified version of invert_ebb to account for sliding windows
[_,_] = invert_sliding_window_ebb(
    sim_fname,
    surf_set,
    patch_size=patch_size,
    n_temp_modes=n_temp_modes,
    n_spatial_modes=n_spatial_modes,
    win_size=25,
    win_overlap=True,
    spm_instance=spm
)
```

Then, we extract the time series and compute the laminar CSD

```python
load_and_plot_csd(sim_fname, surf_set, layer_verts, layer_dists, 'kcsd')
```

Here the spatiotemporal dynamics of layer activity are taken into account.

**Explanation:** Beamformer uses a sensor covariance matrix to compute its inversion. When inverting on the whole window, the covariance contains both layer 3 and layer 9 activity, that have extremely close leadfields, and it collapses into a compromise between the two. It then sees a dominant spatial pattern, for the whole time window. If we invert a sliding time window, we get only one laminar source at a time, spatially decorrelated. But keep in mind that temporal and spatial discrimination of laminar sources depends widely on the width of your reconstructed window. If you take a too small time window, you don't get enough time samples to build an accurate covariance matrix.

## Part II: Correlated sources across layers: the need for specific priors

#### Simulating 2 simultaneous sources

Here we simulate 2 sources in layer 3 and 9 (as before) but co-occuring at the same time.We keep the same SNR = -5bB and dipole moement 5nAm as previously. 

```python
# Strength of simulated activity (nAm)
dipole_moment = 5
# Temporal width of the simulated Gaussian
signal_width= 10 # 25ms
# Sampling rate (must match the data file)
s_rate = 600

zero_time_1new=0
zero_time_2new=0
sim_signal1_new=np.exp(-((time-zero_time_1new)**2)/(2*signal_width**2)).reshape(1,-1)
sim_signal2_new=np.exp(-((time-zero_time_2new)**2)/(2*signal_width**2)).reshape(1,-1)
plt.plot(time,dipole_moment*sim_signal1_new[0,:],label='sim1')
plt.plot(time,dipole_moment*sim_signal2_new[0,:],label='sim2')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (nAm)')
```

### Invert with basic ebb

We use the usual ebb inversion method 

```python
[_,_, MU2] = invert_ebb(
    sim_fname_2,
    surf_set,
    patch_size=patch_size,
    n_temp_modes=n_temp_modes,
    n_spatial_modes=n_spatial_modes,
    return_mu_matrix= True,
    spm_instance=spm
)
```

We can observe that with simultaneous (correlated) sources, we reconstruct a source in the middle point. This issue was already present in the free energy model comparison approach, with the surface in the middle of the two simulated layers being the model with the highest likelihood.

### Invert with ebb_layer

To counteract this behavior which is a common issue with beamformers, we base ourselves on the following implementation of EBB for correlated sources [O’Neill, G.C., Barry, D.N., Tierney, T.M. et al. Testing covariance models for MEG source reconstruction of hippocampal activity. Sci Rep 11, 17615 (2021).](https://doi.org/10.1038/s41598-021-96933-0).
As our mesh has a specific shape of Nvertices x Nlayers, with leadfields within cortical column (1 x Nlayers) highly colinear (linked vectors correspondance), we need to take this structure into account: 

For this 3 priors will be used: 
- an independent prior: a non-zero covariance prior for vertices with the same surface index (but same "cortical column")
- a correlated sum prior: for each layer pair, we create a combined lead field (q+=la+lb)
- a correlated depth contrast prior: for each pair of potential layer sources, we create the difference lead field (q-=la-lb)

```python
# Inverting with this layer ebb is (about 2 times longer than without priors)
[_,_] = invert_sliding_window_ebb_layer(
        sim_fname_2, 
        surf_set, 
        patch_size=patch_size, 
        n_temp_modes=n_temp_modes,
        n_spatial_modes=n_spatial_modes,
        spm_instance=spm
    )
```

```python
load_and_plot_csd(sim_fname_2, surf_set, layer_verts, layer_dists, 'kcsd')
```

EBB layer can then reconstruct 2 simultaneous sources.
